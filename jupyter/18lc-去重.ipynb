4sum问题中有些比较tricky点需要仔细探究，并且在清楚区别后可以发散到大部分的2 pointer的问题中套用。
> 4 sum problem statements:
给你一个由 n 个整数组成的数组 nums ，和一个目标值 target 。请你找出并返回满足下述全部条件且<u>不重复的四元组</u> [nums[a], nums[b], nums[c], nums[d]] （若两个四元组元素一一对应，则认为两个四元组重复）：
`0 <= a, b, c, d < n
a、b、c 和 d 互不相同
nums[a] + nums[b] + nums[c] + nums[d] == target`
你可以按 任意顺序 返回答案 。

###### 1. 理解什么是不重复的四元组
比如在给定nums中我们有 `nums = [1, 0, -1, 0, -2, 2]`, `target = 0` 中， `nums` 排序后得到 `[-2, -1, 0, 0, 1, 2]`。所有符合条件且值各不相同的一一对应组合（distinct quadruplets）为：
* `[-2, -1, 1, 2]`
* `[-2, 0, 0, 2]`
* `[-1, 0, 0, 1]`
这三组四元组在值和顺序上互不相同，视为三种不同的解。
而该nums中由于要实现target 2个0都需要选取所以不会导致重复组合的出现但是其他的case需要对重复值组合进行分析：
比如`nums = [-2, -1, 0, 0, 0, 1, 2] target=0`其中0出现三次，如果不跳过重复组合，双指针会枚举下列索引组合：
- 取 0 对应索引 (2,3)：[-2, 0, 0, 2]
- 取 0 对应索引 (2,4)：[-2, 0, 0, 2]
- 取 0 对应索引 (3,4)：[-2, 0, 0, 2]
结果中会出现3次相同的[-2, 0, 0, 2]。同理[-1,0,0,1]也会出现三次，这样会导致在最终res数组中有重复值组合，不符合题意。所以我们需要跳过相邻相同值来进行去重。
**去重方法是**
```python
# 外层去重
if i>0 and nums[i]==nums[i-1]: continue
if j>i+1 and nums[j]==nums[j-1]: continue
# 内层去重
while l<r and nums[l]==nums[l+1]: l+=1
while l<r and nums[r]==nums[r-1]: r-=1

```
这种需要**去重的重复值情况**叫做distinct value combination,还有其他的重复值要求情况比如 distinct index combination, count combination, subset, subsequence, subarray, permutation等等，维度的不同会导致求解要求差异。还有其他的问题维度比如：
* repetition control（可重复使用 vs 不可重复使用）：如 Combination Sum 允许无限次复用同一元素，而 Combination Sum II 每个元素只能用一次；
* size constraint（固定长度 k 的子集）：如从 n 个元素中选出恰好 k 个子集（k-combinations），需在回溯中维护深度限制；
* target-sum constraint（目标和）：如 Subset Sum（子集和）和 Knapsack（背包）问题，要求选出的组合满足给定和或容量限制；
* order requirement（顺序/字典序）：有的题目要求按 lexicographical order（字典序）输出组合，需要在枚举时保证顺序一致；
  * 字典序处理：通过对输入列表调用 sort() 或 sorted()，并在回溯（backtracking）时按升序遍历候选元素，从而保证生成的排列（permutation）或组合（combination）序列即为字典序（lexicographical order）。
* multi-constraint（多约束）：如同时限制和与长度，或元素之间必须满足差值/距离约束；
* weighted combination（加权组合）：每个元素有不同权重或价值，需要输出满足特定价值区间的组合。



````python
# 5/27 2 pointer solution 
from typing import List
class Solution:
    # 题目4sum要求返回所有不重复且和等于目标值的四元组，若不在枚举过程中跳过相同的nums[i],会多次声称内容相同但索引不同的四元组最终结果会出现重复组合
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        n = len(nums)
        res=[]
        for i in range(n):
            if i>0 and nums[i]==nums[i-1]:
                continue
            for j in range(i+1,n):
                if j>i+1 and nums[j]==nums[j-1]:
                    continue
                l,r = j+1,n-1
                while l<r:
                    t=nums[i]+nums[j]+nums[l]+nums[r]
                    if t==target:
                        res.append([nums[i],nums[j],nums[l],nums[r]])
                        while l<r and nums[l]==nums[l+1]:
                            l+=1
                        while l<r and nums[r]==nums[r-1]:
                            r-=1
                        l+=1
                        r-=1
                    elif t<target:
                        l+=1
                    else:
                        r-=1
        return res
````
